```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Load the wine dataset
wine_dataset = load_wine()

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(wine_dataset.data, wine_dataset.target, test_size=0.5, random_state=42)

# Create a KNN classifier with an optimal number of neighbors (you can tune this)
knn = KNeighborsClassifier(n_neighbors=9)

# Train the KNN classifier on the training data
knn.fit(X_train, y_train)

# Predict the classes on the test data
knn_predicted = knn.predict(X_test)

# Evaluate the model's accuracy on the test data
accuracy = np.mean(knn_predicted == y_test)
print("Accuracy on Test Data: {:.2f}".format(accuracy))
```

```python
y_pred = knn.predict(X_test)
print(format(y_pred))
```
